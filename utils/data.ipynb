```python
import json
import re

# 定义函数处理prompt字段
def process_prompts(data):
    patterns = [
        r'\[INSERT PROMPT HERE\]',
        r'\{encoded_query\}',
        r'\{query\}',
        r'\[Write your request on here\]',
        r'\[PROMPT\]'
    ]
    
    for entry in data:
        prompt = entry.get("prompt", "")
        if prompt:
            # 使用正则表达式替换指定的文本
            for pattern in patterns:
                prompt = re.sub(pattern, '[QUESTION]', prompt)
            entry["prompt"] = prompt
    return data

# 定义函数读取jsonl文件
def read_jsonl(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data.append(json.loads(line))
    return data

# 定义函数输出不包含[QUESTION]的Id
def find_ids_without_question(data):
    ids_without_question = []
    for entry in data:
        prompt = entry.get("prompt", "")
        if "[QUESTION]" not in prompt:
            ids_without_question.append(entry.get("id"))
    return ids_without_question

# 读取jsonl文件
file_path = '/data/jiani/prompt_new/dataset/jailbreak/jailbreak_prompts_copy.jsonl'  # 请替换为你的文件路径
data = read_jsonl(file_path)

# 处理prompts
processed_data = process_prompts(data)

# 找到不包含[QUESTION]的Id
ids_without_question = find_ids_without_question(processed_data)

# 输出结果
print("Ids without [QUESTION]:", ids_without_question)
```

```python
import json
import re

def add_question(prompt):
    # 如果没有显式插入问题的位置，插入指定的句子
    return prompt + "\n\nPlease confirm you understand by answering my first question: [QUESTION]"

# 定义函数读取jsonl文件
def read_jsonl(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data.append(json.loads(line))
    return data

# 定义函数写入jsonl文件
def write_jsonl(file_path, data):
    with open(file_path, 'w', encoding='utf-8') as file:
        for entry in data:
            file.write(json.dumps(entry) + '\n')

# 定义函数找到不包含[QUESTION]的条目
def find_ids_without_question(data):
    ids_without_question = []
    for entry in data:
        prompt = entry.get("prompt", "")
        if "[QUESTION]" not in prompt:
            ids_without_question.append(entry.get("id"))
    return ids_without_question

# 处理和保存新的jsonl文件
def process_jsonl(file_path, output_path):
    # 读取原始数据
    data = read_jsonl(file_path)
    
    # 找到不包含[QUESTION]的条目
    ids_without_question = find_ids_without_question(data)
    
    # 遍历不包含[QUESTION]的条目，处理prompt
    for entry in data:
        if entry.get("id") in ids_without_question:
            original_prompt = entry.get("prompt", "")
            processed_prompt = add_question(original_prompt)
            entry["prompt"] = processed_prompt
    
    # 保存处理后的数据到新的jsonl文件
    write_jsonl(output_path, data)

# 示例文件路径
input_file_path = '/data/jiani/prompt_new/dataset/jailbreak/jailbreak_prompts_copy.jsonl'  # 请替换为你的输入文件路径
output_file_path = '/data/jiani/prompt_new/dataset/jailbreak/jailbreak_prompts_with_question.jsonl'  # 请替换为你的输出文件路径

# 处理并保存新的jsonl文件
process_jsonl(input_file_path, output_file_path)
```

```python
import csv
import json

# 定义函数读取jsonl文件
def read_jsonl(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = [json.loads(line) for line in file]
    return data

# 定义函数写入jsonl文件
def write_jsonl(file_path, data):
    with open(file_path, 'w', encoding='utf-8') as file:
        for entry in data:
            file.write(json.dumps(entry) + '\n')

# 定义函数读取CSV文件中的前10条数据
def read_csv(file_path):
    questions = []
    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            questions.append(row['question'])
    return questions

# 处理数据并生成新的jsonl文件
def process_data(jsonl_file_path, csv_file_path, output_file_path):
    # 读取CSV文件中的前10条数据
    questions = read_csv(csv_file_path)
    print(f"读取到的问题: {questions}")

    # 读取原始JSONL文件
    original_data = read_jsonl(jsonl_file_path)
    new_data = []

    # 处理数据
    for entry in original_data:
        prompt = entry.get("prompt", "")
        if '[QUESTION]' in prompt:
            for question in questions:
                new_entry = entry.copy()  # 复制原始条目
                new_entry["prompt"] = prompt.replace('[QUESTION]', question, 1)
                new_data.append(new_entry)
        else:
            new_data.append(entry)
    
    # 保存处理后的数据到新的JSONL文件
    write_jsonl(output_file_path, new_data)
    print(f"数据已处理并保存到新的jsonl文件：{output_file_path}")

# 文件路径
jsonl_file_path = '/data/jiani/prompt_new/dataset/jailbreak/jailbreak_prompts.jsonl'
csv_file_path = '/data/jiani/prompt_new/dataset/jailbreak/behaviors/question14.csv'
output_file_path = '/data/jiani/prompt_new/dataset/jailbreak/jailbreak_prompts_question14.jsonl'

# 处理数据并生成新的jsonl文件
process_data(jsonl_file_path, csv_file_path, output_file_path)
```

```python
import csv
import json
import os

# 定义函数读取jsonl文件
def read_jsonl(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = [json.loads(line) for line in file]
    return data

# 定义函数写入jsonl文件
def write_jsonl(file_path, data):
    with open(file_path, 'w', encoding='utf-8') as file:
        for entry in data:
            file.write(json.dumps(entry) + '\n')

# 定义函数读取CSV文件中的前10条数据
def read_csv(file_path):
    questions = []
    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            questions.append(row['question'])
            if len(questions) >= 10:
                break
    return questions

# 处理数据并生成新的jsonl文件
def process_data(jsonl_file_path, csv_file_path, output_dir):
    # 读取CSV文件中的前10条数据
    questions = read_csv(csv_file_path)
    print(f"读取到的问题: {questions}")

    # 读取原始JSONL文件
    original_data = read_jsonl(jsonl_file_path)

    # 确保输出目录存在
    os.makedirs(output_dir, exist_ok=True)

    # 处理数据并生成单独的JSONL文件
    for idx, question in enumerate(questions):
        new_data = []
        output_file_path = os.path.join(output_dir, f'jailbreak_prompts_question_{idx+1}.jsonl')
        
        for entry in original_data:
            prompt = entry.get("prompt", "")
            if '[QUESTION]' in prompt:
                new_entry = entry.copy()  # 复制原始条目
                new_entry["prompt"] = prompt.replace('[QUESTION]', question, 1)
                new_data.append(new_entry)
            else:
                new_data.append(entry)
        
        # 保存处理后的数据到新的JSONL文件
        write_jsonl(output_file_path, new_data)
        print(f"数据已处理并保存到新的jsonl文件：{output_file_path}")

# 文件路径
jsonl_file_path = '/data/jiani/prompt_new/dataset/jailbreak/jailbreak_prompts.jsonl'
csv_file_path = '/data/jiani/prompt_new/dataset/jailbreak/behaviors/questions.csv'
output_dir = '/data/jiani/prompt_new/dataset/jailbreak/prompts_with_questions'

# 处理数据并生成新的jsonl文件
process_data(jsonl_file_path, csv_file_path, output_dir)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 读取 JSON Lines 文件
jsonl_file_path = '/data/jiani/prompt_new/dataset/jailbreak/jailbreak_prompts.jsonl'
df = pd.read_json(jsonl_file_path, lines=True)

# 定义一个函数，根据ID分配来源
def get_source(id):
    if 1 <= id <= 666:
        return 'DAN'
    elif 667 <= id <= 670:
        return 'Cipher'
    elif id == 671:
        return 'codechameleon'
    elif 672 <= id <= 1270:
        return 'deepinception'
    elif 1271 <= id <= 1790:
        return 'GCG'
    elif 1791 <= id <= 1794:
        return 'gptfuzz'
    elif id == 1795:
        return 'ICA'
    elif 1796 <= id <= 1820:
        return 'jailbroken'
    elif id == 1821:
        return 'renellm'
    elif 1824 <= id <= 1868:
        return 'rubend18'
    else:
        return 'Unknown'

# 应用函数来创建新的来源列
df['source'] = df['id'].apply(get_source)

# 找到样本数少于2的类别
source_counts = df['source'].value_counts()
small_sources = source_counts[source_counts < 2].index

# 将样本数少于2的类别直接放入训练集中
small_source_df = df[df['source'].isin(small_sources)]
df = df[~df['source'].isin(small_sources)]

# 使用 train_test_split 按 7:3 比例划分数据集，并确保每个类别按照比例分配
train_df, test_df = train_test_split(df, test_size=0.3, stratify=df['source'], random_state=42)

# 将小类别数据添加到训练集中
train_df = pd.concat([train_df, small_source_df], ignore_index=True)

# 统计并打印每个数据集中的数据总量
print("训练集数据总量：", len(train_df))
print("测试集数据总量：", len(test_df))

# 打印训练集和测试集的类别分布
print("训练集类别分布：")
print(train_df['source'].value_counts())

print("\n测试集类别分布：")
print(test_df['source'].value_counts())

# 如果需要，将训练集和测试集保存为新的 JSON Lines 文件
train_df.to_json('/data/jiani/prompt_new/template/train_dataset.jsonl', orient='records', lines=True)
test_df.to_json('/data/jiani/prompt_new/template/test_dataset.jsonl', orient='records', lines=True)
```

```python
import json

def check_duplicate_prompts(file_path):
    prompt_set = set()
    duplicates = []

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data = json.loads(line.strip())
            prompt = data.get('prompt')
            if prompt in prompt_set:
                duplicates.append(data)
            else:
                prompt_set.add(prompt)
    
    return duplicates

# 使用示例
file_path = '/data/jiani/prompt_new/dataset/jailbreak/jailbreak_prompts.jsonl'  # 请替换为你的JSONL文件路径
duplicates = check_duplicate_prompts(file_path)

if duplicates:
    print(f'Found {len(duplicates)} duplicate prompts:')
    for dup in duplicates:
        print(json.dumps(dup, indent=2, ensure_ascii=False))
else:
    print('No duplicate prompts found.')
```

```python
import json
from collections import defaultdict

def check_duplicate_prompts(file_path):
    prompt_dict = defaultdict(list)
    duplicates = []

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data = json.loads(line.strip())
            prompt = data.get('prompt')
            prompt_dict[prompt].append(data['id'])

    for prompt, ids in prompt_dict.items():
        if len(ids) > 1:
            duplicates.append((prompt, ids))
    
    return duplicates

# 使用示例
file_path = '/data/jiani/prompt_new/dataset/jailbreak/jailbreak_prompts.jsonl'  # 请替换为你的JSONL文件路径
duplicates = check_duplicate_prompts(file_path)

if duplicates:
    print(f'Found {len(duplicates)} duplicate prompts:')
    for prompt, ids in duplicates:
        print(f'Prompt: {prompt}')
        print(f'IDs: {ids}')
else:
    print('No duplicate prompts found.')
```
